```python
import pandas as pd
import numpy as np

interesting_columns = [
    'latitude',
    'longitude',
    'housing_median_age',
    'total_rooms',
    'total_bedrooms',
    'population',
    'households',
    'median_income',
    'median_house_value',
    'ocean_proximity',
]

df = pd.read_csv('housing.csv')
df = df[interesting_columns].fillna(0)
df['rooms_per_household'] = df.total_rooms/df.households
df['bedrooms_per_room'] = df.total_bedrooms/df.total_rooms
df['population_per_household'] = df.population/df.households
df.head().T
```

#### Q2: Create the correlation matrix and find the two features with the biggest correlation?

```python
from IPython.display import display
numerical = df.columns[df.dtypes != 'object'].to_list()
corr_M = pd.DataFrame(index=numerical, columns=numerical)
for column_a in numerical:
    for column_b in numerical:
        corr_M.loc[column_a, column_b] = df[column_a].corr(df[column_b])
        


display(corr_M)
```

```python
max_correlation = 0
correlated_ones = None
for i in range(len(numerical)):
    for j in range(i+1, len(numerical)):
        if corr_M.loc[numerical[i], numerical[j]] > max_correlation:
            max_correlation = corr_M.loc[numerical[i], numerical[j]]
            correlated_ones = (numerical[i], numerical[j])
print(max_correlation, correlated_ones)
```

#### Make median_house_value binary

```python
m = df.median_house_value.mean()
df['above_average'] = (df['median_house_value'] > m).astype(int)
```

```python

from sklearn.model_selection import train_test_split

np.random.seed(42)

df_full_train, df_test = train_test_split(df, test_size=0.2)

df_train, df_val = train_test_split(df_full_train, test_size=0.25)

y_test = df_test.above_average.to_numpy()
yp_test = df_test.median_house_value.to_numpy()
del df_test['above_average']
del df_test['median_house_value']

y_val = df_val.above_average.to_numpy()
yp_val = df_val.median_house_value.to_numpy()
del df_val['above_average']
del df_val['median_house_value']

y_train = df_train.above_average.to_numpy()
yp_train = df_train.median_house_value.to_numpy()
del df_train['above_average']
del df_train['median_house_value']

```

#### Q3: Mutual information

```python
from sklearn.metrics import mutual_info_score
mi_score = mutual_info_score(df['ocean_proximity'], df['above_average'])
np.round(mi_score, decimals=2)
```

#### Q4: Train logistic regression

```python
df.ocean_proximity.unique()
```

```python
from sklearn.feature_extraction import DictVectorizer

dv = DictVectorizer(sparse=False)

train_dict = df_train.to_dict(orient='records')
X_train = dv.fit_transform(train_dict)

val_dict = df_val.to_dict(orient='records')
X_val = dv.transform(val_dict)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(solver="liblinear", C=1.0, max_iter=1000, random_state=42)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_val)
global_accuracy = (y_pred == y_val).mean()
np.round(global_accuracy, decimals=2)
```

#### Q5: find the least useful feature using the feature elimination technique.

```python

q5_columns = ['total_rooms', 'total_bedrooms', 'population', 'households']


dv = DictVectorizer(sparse=False)

train_dict = df_train[q5_columns].to_dict(orient='records')
_X_train = dv.fit_transform(train_dict)

val_dict = df_val[q5_columns].to_dict(orient='records')
_X_val = dv.transform(val_dict)

model = LogisticRegression(solver="liblinear", C=1.0, max_iter=1000, random_state=42)
model.fit(_X_train, y_train)

y_pred = model.predict(_X_val)
_accuracy = (y_pred == y_val).mean()


less_important_feature = None
best_diff = 1
for column in q5_columns:
    dv = DictVectorizer(sparse=False)

    train_dict = df_train[q5_columns].drop(labels=column, axis=1).to_dict(orient='records')
    _X_train = dv.fit_transform(train_dict)

    val_dict = df_val[q5_columns].drop(labels=column, axis=1).to_dict(orient='records')
    _X_val = dv.transform(val_dict)

    model = LogisticRegression(solver="liblinear", C=1.0, max_iter=1000, random_state=42)
    model.fit(_X_train, y_train)

    y_pred = model.predict(_X_val)
    current_accuracy = (y_pred == y_val).mean()
    if np.abs(current_accuracy-_accuracy) < best_diff:
        best_diff = np.abs(current_accuracy-_accuracy)
        less_important_feature = column
display(less_important_feature, best_diff)
```
